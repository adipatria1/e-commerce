```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load datasets
customers = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/customers_dataset.csv')
geolocation = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/geolocation_dataset.csv')
order_items = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/order_items_dataset.csv')
orders = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/orders_dataset.csv')
order_reviews = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/order_reviews_dataset.csv')
order_payments = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/order_payments_dataset.csv')
product_category_name_translation = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/product_category_name_translation.csv')
products = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/products_dataset.csv')
sellers = pd.read_csv('/content/drive/MyDrive/Datasets/ecommerce/sellers_dataset.csv')
```

```python
# drop rows with null values
customers.dropna(inplace=True)
geolocation.dropna(inplace=True)
order_items.dropna(inplace=True)
orders.dropna(inplace=True)
order_reviews.dropna(inplace=True)
order_payments.dropna(inplace=True)
product_category_name_translation.dropna(inplace=True)
products.dropna(inplace=True)
sellers.dropna(inplace=True)

# print cleaned datasets
print(customers.head())
print(geolocation.head())
print(order_items.head())
print(orders.head())
print(order_reviews.head())
print(order_payments.head())
print(product_category_name_translation.head())
print(products.head())
print(sellers.head())
```

```python
print(customers.duplicated().sum())
print(geolocation.duplicated().sum())
print(order_items.duplicated().sum())
print(orders.duplicated().sum())
print(order_reviews.duplicated().sum())
print(order_payments.duplicated().sum())
print(product_category_name_translation.duplicated().sum())
print(products.duplicated().sum())
print(sellers.duplicated().sum())
```

```python
customers.drop_duplicates(inplace=True)
geolocation.drop_duplicates(inplace=True)
order_items.drop_duplicates(inplace=True)
orders.drop_duplicates(inplace=True)
order_reviews.drop_duplicates(inplace=True)
order_payments.drop_duplicates(inplace=True)
product_category_name_translation.drop_duplicates(inplace=True)
products.drop_duplicates(inplace=True)
sellers.drop_duplicates(inplace=True)
```

```python
customers.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_customers.csv', index=False)
geolocation.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_geolocation.csv', index=False)
order_items.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_order_items.csv', index=False)
orders.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_orders.csv', index=False)
order_reviews.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_order_reviews.csv', index=False)
order_payments.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_order_payments.csv', index=False)
product_category_name_translation.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_product_category.csv', index=False)
products.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_products.csv', index=False)
sellers.to_csv('/content/drive/MyDrive/Datasets/ecommerce/cleaned_sellers.csv', index=False)
```
